```python
import pandas as pd
import numpy as np
import os
Target_data = r'F:\Integrated_analysis_data\Data\result.xls'
Join_data = r'F:\Integrated_analysis_data\Data\副本npp_站点2.xls'
outpath = r'F:\Integrated_analysis_data\Data'
from IPython.display import display
```

```python
Target = pd.read_excel(Target_data)
# display(Target)
# Join_soc = pd.read_excel(Join_data,sheet_name='土壤有机碳密度')
# display(Join_soc)
Join_up = pd.read_excel(Join_data,sheet_name='植被地上碳密度')
# display(Join_up)
Join_down = pd.read_excel(Join_data,sheet_name='植被地下碳密度')
# display(Join_down)
Target_join_data_up = pd.merge(Target,Join_up,on = ['Longitude','Latitude'],how = 'left')
display(Target_join_data_up)
norepeat_df = Target_join_data_up.drop_duplicates(subset=['Code_x'], keep='first')
display(norepeat_df)
Target_join_data_down = pd.merge(norepeat_df,Join_down,on = ['Longitude','Latitude'],how = 'left')
display(Target_join_data_down)
norepeat_df = Target_join_data_down.drop_duplicates(subset=['Code_x'], keep='first')
display(norepeat_df)
Target_join_data_down = norepeat_df.dropna(axis=0, how='any')
result = Target_join_data_down.reset_index(drop=True)
display(result)
result.to_excel(outpath + os.sep + 'result_join.xls')
```
