```python
import ee
import geemap
Map = geemap.Map()

geometry = ee.Geometry.Point([-104.78007812499999, 37.07682857623016])
l8Col = ee.ImageCollection("LANDSAT/LC08/C01/T1_RT_TOA") \
              .filterDate("2018-2-1", "2018-5-1") \
              .filterBounds(geometry)

def func_jdg(image):
                return image.addBands(image.normalizedDifference(["B5", "B4"]).rename("NDVI")) \
              .map(func_jdg) \
              .select("NDVI")
print("l8Col", l8Col)

def func_xyd(image):
    time_start = image.get("system:time_start")
    year = ee.Date(time_start).get("year")
    image = image.set("year", year)
    return image

newCol = l8Col.map(func_xyd) \
.aside(print) \
.aside(Map.addLayer, {min:0, max:0.8, palette:"ff0000,00ff00,0000ff"}, "image")
MapMap
```
