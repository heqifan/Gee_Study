```python

# set study area
Ukraine = ee.FeatureCollection("FAO/GAUL/2015/level0") \
                .filter(ee.Filter.eq('ADM0_NAME','Ukraine'))
roi = Ukraine
Map.addLayer(roi, {'color':'grey'}, 'studyArea')
Map.centerObject(roi)

# Create an example feature collection of points
exampleFC = ee.FeatureCollection.randomPoints(roi, 100).randomColumn('PoI')
Map.addLayer(exampleFC, {'color':'red'}, 'exampleFC')
print("exampleFC",exampleFC.getInfo())

# k folder value
k = 10

def func_zgq(f):
    return f.set('fold', ee.Number(f.get('random')).multiply(k).floor())

collection = exampleFC.randomColumn('random').map(func_zgq)

folds = collection.distinct('fold').select('fold')
print('folds:',folds)

filter = ee.Filter.equals({'leftField': 'fold', 'rightField':'fold'})
folds = ee.Join.saveAll('training').apply(folds, collection, filter.Not())
folds = ee.Join.saveAll('validation').apply(folds, collection, filter)

print("folds",folds)

```
